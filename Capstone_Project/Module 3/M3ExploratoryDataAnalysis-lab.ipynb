## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
plt.figure(figsize=(10,5))
plt.boxplot(comp.describe(), vert=False)
plt.xlabel("Compensation")
plt.ylabel("Data")
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
quartile_vals = comp.quantile([0.25, 0.75])
iqr = quartile_vals[0.75] - quartile_vals[0.25]
print("Inter quartile range: ", iqr)
print("Median: ", comp.quantile([0.5]))
```

Find out the upper and lower bounds.


```python
# your code goes here
upper_bound = quartile_vals[0.75] + 1.5*iqr
lower_bound = quartile_vals[0.25] - 1.5*iqr
print("Upper bound: ", upper_bound)
print("Lower bound: ", lower_bound)
type(upper_bound)
type(df['ConvertedComp'])
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
# to find the numeber of outliers, find all entries that are above or below the bound values
import math
df_comp_list = list(comp)
df_comp_list = [x for x in comp if not math.isnan(x)]
outlier_count = 0 
for entry in df_comp_list:
    if entry > upper_bound or entry < lower_bound:
        outlier_count += 1
print("Number of outliers: ", outlier_count)
```

```python
comp_out_upp = comp[lambda x : x > upper_bound]
comp_out_low = comp[lambda x : x < lower_bound]
comp_out_upp
```

```python
print("Median of compensation after removi")
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
# now all of the outliers are above the upper bound
df_no_out = df[df['ConvertedComp'] < upper_bound]
df_no_out.head()
```

```python
print("Median after removing outliers: ", df_no_out['ConvertedComp'].median())
print("Mean after removing outliers: ",df_no_out['ConvertedComp'].mean())
```
